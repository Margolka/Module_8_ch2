```python
import pandas as pd
data = pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/', header=0)
df=data[0]
```

1. Zmiana nagłówków kolumn na poslkie odpowiedniki.

```python
df.rename(columns={'TITLE':'TYTUŁ','ARTIST':'ARTYSTA','YEAR':'ROK','HIGH POSN':'MAX POZ'}, inplace=True)
df.head()
```

2. Liczba pojedynczy artystó na liśnine:

```python
df['ARTYSTA'].nunique()
```

3. Najczęściej pojawiający się na liście: Coldplay i Take That

```python
df['ARTYSTA'].value_counts().head()
```

4. Zmiana nagłówków

```python
df.rename(str.capitalize, axis='columns', inplace=True)
df.head()
```

5. Wyrzucamy kolumnę 'Max Poz'

```python
df.drop('Max poz',axis=1,inplace=True)
df.head()
```

6. Najwięcej alabumów wyszło w latach: 2000, 1987

```python
df.groupby('Rok').count().sort_values(by='Tytuł',ascending=False).head()
```

7. W okresie 1960-1990 z albumów znajdujących się na liście wydano 22 albumy.

```python
df[(df['Rok'] >= 1960) & (df['Rok'] <= 1990)].count()[0]
```

8. Najmłodszy album na liście został wydany w:

```python
df.sort_values(by='Rok', ascending=False).head(1)
```

9. Lista najwcześniej wydanych albumów każdego artysty, który znalazł się na liście.

```python
df.groupby(['Artysta','Tytuł']).apply(lambda x: x.sort_values(by='Rok', ascending=False))
```
